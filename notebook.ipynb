```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
for lines in data['DESCR'].split('\n'):
    print(lines)
```

```python
x = data['data']

y = data['target']
```

```python
data.feature_names
```

```python
df = pd.DataFrame(x,columns = data.feature_names)
df.head(10)
```

```python
df['Class'] = y
df.head(10)
```

```python
for i in df.columns:
    q1 = np.quantile(df[i],0.25)
    q3 = np.quantile(df[i],0.75)

    iqr = q3-q1

    lb = q1 -  iqr*1.5
    ub = q3 +  iqr*1.5

    outlair = np.where(((df[i]<lb)|(df[i]>ub)))
    for index in outlair[0]:
        df[i].iloc[index]=None

df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
x = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split as spliter

x_train,x_test,y_train,y_test  = spliter(x,y,train_size=0.8,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(x_train,y_train)
```

```python
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score

print("Training Evaluation")
print(confusion_matrix(y_train,y_train_pred))
print(accuracy_score(y_train,y_train_pred))
```

```python
print("Testing Evaluation")
print(confusion_matrix(y_test,y_test_pred))
print(accuracy_score(y_test,y_test_pred))
```

```python
import pickle

pickle.dump(model,open('iris_model.pkl','wb'))
```
